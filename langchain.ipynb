```python
import pandas as pd
column_names=["category","description"]
df = pd.read_csv("/content/ecommerceDataset.csv",names=column_names,header=None)
df.shape
```

```python
df
```

```python
categories = ['Household', 'Books', 'Clothing & Accessories', 'Electronics']

# Create an empty dataframe to store the sampled data
sampled_df = pd.DataFrame()

# Sample 5000 entries for each category
for category in categories:
    category_df = df[df['category'] == category]
    sampled_category_df = category_df.sample(n=5000, random_state=1)
    sampled_df = pd.concat([sampled_df, sampled_category_df])

# Reset the index of the sampled dataframe
sampled_df.reset_index(drop=True, inplace=True)

# Display the shape of the sampled dataframe
print(sampled_df.shape)
```

```python
df.category.unique()
```

```python
sampled_df
```

```python
sampled_df.to_csv("/content/sampled_ecommerceDataset.csv", index=False)
```

```python
from sentence_transformers import SentenceTransformer
encoder=SentenceTransformer("all-mpnet-base-v2")
vectors=encoder.encode(sampled_df.description)
```
